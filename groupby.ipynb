```python
import pandas as pd
```

```python
pokemons = pd.read_csv('Pokemon.csv', index_col='#')
pokemons
```

```python
#get nubers for each group
pokemons['Type 1'].value_counts()
```

```python
#group values by column
poke_groups = pokemons.groupby(by='Type 1')
```

```python
#return groups with indicies as a value
poke_groups.groups
```

```python
#get group by refering to type
poke_groups.get_group('Grass').head(10)
```

```python
#only need to create group once to reuse
poke_groups.get_group('Fire').head(10)
```

```python
#its easier to perform pivot_tables with groups
#pokemons.pivot_table(index='Type 1', aggfunc='mean').round(2)
```

```python
poke_groups.mean(numeric_only=True).round(2)
```

```python
#return specific data
poke_groups.get_group('Grass').mean(numeric_only=True).round(2)
```

```python
#groupby wih more than one column
poke_groups =pokemons.groupby(by=['Type 1', 'Type 2'], dropna=False)
poke_groups
```

```python
#return groups with indicies as a value with 2 elements as key
poke_groups.groups
```

```python
#using groups to perform operations
poke_max = poke_groups.max(),round(2)
poke_max
```

```python
#get values for subtype
#poke_max.loc['Grass'].sort_values(by='Total', ascending=False)
```

```python
#using the agg function with groups
pokemons.groupby('Type 1').agg({'Total': ['max', 'min', 'mean']}).round(2)
```

```python
#with uliple groups
poke_groups = pokemons.groupby(by=['Type 1', 'Type 2'], dropna=False)
poke_groups
```

```python
poke_groups.agg({'Total': ['max', 'min','mean']}).round(2)
```
