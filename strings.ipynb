```python
import pandas as pd
```

```python
pokemons = pd.read_csv('Pokemon.csv', index_col='#')
pokemons
```

```python
#.str notation needs to be used to convert everythig to string before any opertauions, even strings need converting as 
# pandas see them as objects, only works on string values
pokemons['Name'].str.upper()
```

```python
pokemons['Name'].str.lower()
```

```python
#reverse strings
pokemons['Name'].str[::-1]
```

```python
#filtering on contains a given letter, ignoring case or not
x_filter = pokemons['Name'].str.contains('q', case=False)
pokemons[x_filter]
```

```python
#filter is starts with, is case sensitive
name_filter = pokemons['Name'].str.startswith('Do')
pokemons[name_filter]
```

```python
#filter ends with, case sensitive
name_filter = pokemons['Name'].str.endswith('chu')
pokemons[name_filter]
```

```python
# .replace() to replace values
pokemons['Type 1'].str.replace('Grass', 'Vegan', regex=False)
```

```python
#assig to column
pokemons['Type 1'] = pokemons['Type 1'].str.replace('Grass', 'Vegan', regex=False)
pokemons[pokemons['Type 1'] == 'Vegan']
```

```python
#Using .str on columns
pokemons.columns.str.upper()
```

```python
pokemons.columns = pokemons.columns.str.upper()
pokemons
```
