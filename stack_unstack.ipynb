```python
import pandas as pd
```

```python
rich = pd.read_csv('TopRichestInWorld.csv', index_col='Name')
rich
```

```python
#stacking a dataset
#stacks all column names and uses them as inner most index
#creates inner index for each
rich = rich.stack(dropna=True)
rich
```

```python
pd.DataFrame(rich)
```

```python
#unstacking
rich = rich.unstack()
rich
```
