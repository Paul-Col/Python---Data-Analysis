```python
import pandas as pd
```

```python
pokemons = pd.read_csv('Pokemon.csv')
pokemons
```

```python
#set index
pokemons.set_index(['Name', 'Type 1'], inplace=True)
pokemons
```

```python
#returns index as a tuple
pokemons.index
```

```python
#use get_level_values method 
pokemons.index.get_level_values(1)
```

```python
#swap level method, swap index order
pokemons = pokemons.swaplevel()
pokemons
```

```python
#sort values by Type 1
pokemons.sort_values('Type 1')
```

```python
#select all grass pokemons
pokemons.loc['Grass']
```

```python
#select sepcific columns only
pokemons.loc['Grass', 'Attack']
```

```python
#return from multiple indexes
pokemons.loc[('Grass', 'Bulbasaur')]
```

```python
#return seleced columns for multiple index
pokemons.loc[('Grass', 'Bulbasaur'), ['Attack', 'Defense']]
```

```python
#return seleced columns for multiple index
pokemons.loc[('Grass', 'Bulbasaur'), 'Attack']
```

```python
#using muliple indicies 
pokemons.loc[(['Grass', 'Fire'],['Bulbasaur', 'Litleo']), :]
```
