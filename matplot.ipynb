```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#create sample data
cols = ['open', 'close', 'date']

apple_stocks = pd.read_csv('AAPL.csv', usecols=cols, index_col='date', parse_dates=['date'])
apple_stocks
```

```python
#create sample plot
apple_stocks.plot()
```

```python
#import data to create line plot
cols = ['date', 'high', 'low', 'open', 'close']

ap_stocks = pd.read_csv('AAPL.csv', usecols=cols, index_col='date', parse_dates=['date'])
ap_stocks
```

```python
#create plot showing all data for given month
ap_stocks.loc['2016-05'].plot()
```

```python
#plot for close price only
ap_stocks.plot(y='close')
```

```python
#plot for open and close price in given month
ap_stocks.loc['2016-05'].plot(y=['open', 'close'])
```

```python
#create subplots for each
ap_stocks.plot(subplots=True)
```

```python
#use figsize to change plot size
ap_stocks.plot(subplots=True, figsize=(15,10))
```

```python
#add labels to plot
ap_stocks.plot()
plt.title('Apple Stock Price')
plt.legend(loc='best')
plt.xlabel('Years')
plt.ylabel('Price')
plt.grid()
```

```python
ap_stocks.loc['2016-05'].plot(y='close', c='g', linewidth=5, linestyle=':')
```

```python
#import data to create bar plot
cols = ['date', 'high', 'low', 'open', 'close']

app_stocks = pd.read_csv('AAPL.csv', usecols=cols, index_col='date', parse_dates=['date'])
app_stocks
```

```python
#get the mean
months = app_stocks.resample('a').mean().round(2)
months.index = months.index.year
months
```

```python
#plot bar graph
months.plot(y='high', kind='bar')
plt.xlabel('Year')
```

```python
#turn bars horizontal
months.plot(y='high', kind='barh')
plt.xlabel('year')
```

```python
#multiple bars and changing colours
months.plot(y=['high', 'low'], kind='bar', color=('orange', 'green'))
plt.xlabel('Year')
```

```python
#import data to create histograms
exams = pd.read_csv('exams.csv')
exams
```

```python
#create total column
exams['total'] = exams['math score'] + exams['reading score'] + exams['writing score']
exams
```

```python
#create histogram of total
exams.plot(y='total', kind='hist')
```

```python
#change number of bins and change colour
exams.plot(y='total', kind='hist', bins=25, color='purple')
```

```python
exams['total'].hist(grid=True, bins=20, color='skyblue')
```

```python
#import data to create piecharts
rich = pd.read_csv('TopRichestInWorld.csv')
rich
```

```python
#create counts of each industry
industry = rich['Industry'].value_counts()
industry
```

```python
#plot pie chart
plt.pie(x=industry.values, labels=industry.index)
```

```python
#display only top 10 industries
labels = industry.index[:10]
values = industry.values[:10]

plt.pie(x=values, labels=labels, autopct='%.2f%%')
plt.show()
```

```python
#specify colours, if not enough it will loop through
plt.figure()
plt.pie(x=values, labels=labels, colors=('skyblue','darkblue'))
```
