```python
import pandas as pd
```

```python
pokemons = pd.read_csv('Pokemon.csv', index_col='#')
pokemons
```

```python
pokemons.info()
```

```python
#get info on dataset
pokemons.describe().round()
```

```python
#describe the info that is not numeric
pokemons.describe(exclude=['int', 'float'])
```

```python
#create column for length of name
pokemons['name_length'] = pokemons['Name'].apply(len)
pokemons
```

```python
#delete name length coloumn
del pokemons['name_length']
pokemons
```

```python
#function to determine strength
def strength(attack: int) -> str:
    if attack > 150:
        return 'Strong'
    elif attack > 100:
        return 'Average'
    else:
        return 'Weak'
```

```python
#populate strength column calling above function using .apply()
pokemons['Strength'] = pokemons['Attack'].apply(strength)
pokemons
```

```python
pokemons.value_counts('Strength')
```

```python
#.apply() using multiple columns
def score(row):
    attack: int = row['Attack']
    defense: int = row['Defense']
    speed: int = row['Speed']

    return sum([attack, defense, speed])
```

```python
#create new column using above function
pokemons['Score'] = pokemons.apply(score, axis=1)
pokemons
```

```python
#using lamdas with apply, shortens each name to 5 chars
short_name = pokemons['Name'].apply(lambda x: x[:5])
pokemons.insert(1, 'Short', short_name)
pokemons
```

```python
#map(), maps series, replace with new values
pokemons['Legendary'] = pokemons['Legendary'].map({False: 'No', True: 'True'})
pokemons
```

```python
#not specifying all values will cause the not specifoed to become NaN
pokemons['Generation'].map({1: 'First'})
```

```python
#applymap(), apply funcion throughout dataframe
pokemons.applymap(type)
```

```python
#use applymap to change all values o strings nd apply uppercase
def uppercase(var: str) -> str:
    return var.upper()
```

```python
pokemons = pokemons.applymap(str)
pokemons.applymap(uppercase)
#pokemons.info()
```

```python
#get length of all strings
pokemons.applymap(len)
```

```python
#use astype() to convert data types, NaN cant be converted to numbers
#these examples change form int64 to smaller more appropriate types saving memory
pokemons['Generation'] = pokemons['Generation'].astype('int8')
pokemons['Total'] = pokemons['Total'].astype('int16')
pokemons.info()
```

```python
#change True and False to 1 and 0
pokemons['Legendary'] = pokemons['Legendary'].astype('int8')
pokemons
```

```python
pokemons.info
```

```python
#replacing single values using .replace()
pokemons['Legendary'].replace({True : 'Yes', False : 'No'}, inplace=True)
pokemons
```

```python
#only replace one value
pokemons['Generation'].replace({1 : 'First'}, inplace=True)
pokemons
```

```python
#use .replace() on entire dataframe
pokemons.replace({'Grass': '!', 'First' : '!!', 'Yes' : '!!!'})
```

```python
#use .where(), if condition false reterns False
pokemons.where(pokemons['Generation'] == 2)
```

```python
#use dropna to show filtered
pokemons.where(pokemons['Generation'] == 2).dropna()
```

```python
#use where to show placeholder where it doent match criteria, use dropna to remove
pokemons = pokemons.where(pokemons['Type 1'] == 'Grass', other=None)
pokemons.dropna()
```

```python
#use agg() to aggregate one or more opertions
pokemons.agg({'Attack' : 'mean', 'Defense' : 'max'})
```

```python
 #specify muliple agg
pokemons.agg({'Attack': ['min', 'max', 'std'], 'Defense': ['min', 'max', 'mean']})
```

```python
#return max values for all numeric
pokemons.agg('max', numeric_only=True)
```

```python
#specify multiple operations
pokemons.agg(['max','min'], numeric_only=True)
```
