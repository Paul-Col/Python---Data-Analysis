```python
import pandas as pd
```

```python
com21 = pd.read_csv('company21.csv', sep=';')
com21
```

```python
com22 = pd.read_csv('company22.csv', sep=';')
com22
```

```python
#outer join, will join everything
pd.merge(left=com21,
         right=com22,
         how='outer',
         on='Name',
         suffixes=('_21', '_22'),
         indicator=True)
```

```python
#inner join, only data that is in both is used
pd.merge(left=com21,
         right=com22,
         how='inner',
         on='Name',
         suffixes=('_21', '_22'),
         indicator=True)
```

```python
#left join, returns everything form left dataset, intersection included
pd.merge(left=com21,
         right=com22,
         how='left',
         on='Name',
         suffixes=('_21', '_22'),
         indicator=True)
```

```python
#right join, returns everything form right dataset, intersection included
pd.merge(left=com21,
         right=com22,
         how='right',
         on='Name',
         suffixes=('_21', '_22'),
         indicator=True)
```

```python
#left minus join, you need to create a mask, same for right minus
merged = pd.merge(left=com21,
                  right=com22,
                  how='left',
                  on='Name',
                  suffixes=('_21','_22'),
                  indicator=True)

merged[merged['_merge'] == 'left_only']
```

```python
#outer join minus, everthing from left and right that but not in both
merged = pd.merge(left=com21,
                  right=com22,
                  how='outer',
                  on='Name',
                  suffixes=('_21','_22'),
                  indicator=True)

merged[merged['_merge'] != 'both']
```

```python
#join with different column names
#create new dataset with names changed to employee
com23 = pd.read_csv('company22.csv', sep=';')
com23.rename(columns={'Name': 'Employee'}, inplace=True)
com23
```

```python
#merge on different column names
pd.merge(left=com21,
         right=com23,
         how='outer',
         left_on='Name',
         right_on='Employee',
         suffixes=('_21', '_22'),
         indicator=True)
```
