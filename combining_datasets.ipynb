```python
import pandas as pd
```

```python
data = {'A': [1,2,3],
        'B': [4,5,6],
        'C': [7,8,9]}

data2 = {'A': [10,11,12],
         'B': [13,14,15],
         'C': [16,17,18],
         'D': [19,20,21]}

df1 = pd.DataFrame(data)
df2 = pd.DataFrame(data2)
```

```python
#concat datasets
pd.concat([df1,df2])
```

```python
#use axis=1 for horizontal
pd.concat([df1,df2], axis=1)
```

```python
#give keys to datset
pd.concat([df1,df2], keys=['Data1', 'Data2'])
```

```python
#give keys horizontal
pd.concat([df1,df2], keys=['Data 1', 'Data 2'], axis=1)
```

```python
#ignore index, will reset index and start form 0
pd.concat([df1,df2], ignore_index=True)
```

```python
#with differet indices
data1 = {'A': [1,2,3],
        'B': [4,5,6],
        'C': [7,8,9]}

data2 = {'A': [10,11,12],
         'B': [13,14,15],
         'C': [16,17,18],
         'D': [19,20,21]}

df3 = pd.DataFrame(data, index=[1,2,3])
df4 = pd.DataFrame(data2, index=[4,5,6])
```

```python
#basic concat has no effect
pd.concat([df3,df4])
```

```python
#setting axis=1 has effect filling with NaN
pd.concat([df3, df4], axis=1)
```

```python
#with different column names
data3 = {'A': [1,2,3],
        'B': [4,5,6],
        'C': [7,8,9]}

data4 = {'D': [10,11,12],
         'E': [13,14,15],
         'F': [16,17,18],
         'G': [19,20,21]}

df5 = pd.DataFrame(data3)
df6 = pd.DataFrame(data4)
```
