```python
import pandas as pd
```

```python
cols = ['Series_Title', 'IMDB_Rating', 'No_of_Votes', 'Gross', 'Released_Year', 'Genre']
```

```python
movies = pd.read_csv('imdb_top_1000.csv', usecols=cols)
movies
```

```python
movies.info()
```

```python
movies.describe().round()
```

```python
movies['Gross'] = movies['Gross'].str.replace(',','') #removes commas from Gross string
movies
```

```python
movies['Gross'] = movies['Gross'].astype('int64') #overides colun with new int value
movies
```

```python
movies.describe().round()
```

```python
movies.count()
```

```python
movies.mean(numeric_only=True).round()
```

```python
movies.max(numeric_only=True).round()
```

```python
movies.min(numeric_only=True).round()
```
