```python
cols = ['name','fare','survived','pclass','age','sex']
element = pd.read_excel('titanic3.xlsx', index_col='name', usecols=cols)
element
```

```python
#reduce every fare by half
element['fare'] / 2
```

```python
#create new column for the reduced fare
element['reduced'] = (element['fare'] / 2).round()
element
```

```python
#find all males
element['sex'] == 'male'
```

```python
#filter all males
element[element['sex'] == 'male']
```

```python
#create mask to make it more readable
male_mask = element['sex'] == 'male'
element[male_mask]
```

```python
#create a mask for aged over 65
age_mask = element['age'] > 65
element[age_mask]
```

```python
#filter all females in 1st class that survived
female_mask = element['sex'] == 'female'
first_mask = element['pclass'] == 1
survived_mask = element['survived'] ==1

element[female_mask & first_mask & survived_mask]
```

```python
#filter anyone under age of 21 and paid a fare of more than 200
richchild_mask  = (element['age'] < 21) & (element['fare'] > 200)
rich = element[richchild_mask]

#find top five
rich.nlargest(n=5, columns=['age','fare'])
```

```python
#filter everyone under age of 21 OR female and survived
survived_mask = element['survived'] == 1
femalke_mask = element['sex'] == 'female'
age_mask = element['age'] <= 21

element[(age_mask | female_mask) & survived_mask]
```

```python
#filter people between 20 and 25
upper_mask = element['age'] >= 20
lower_mask = element['age'] <= 25

element[upper_mask & lower_mask]
```

```python
#easier way is to use between method
youngadult_mask = element['age'].between(20,25)
element[youngadult_mask]
```

```python
#select passengers in 1st or 2nd class
first_mask = element['pclass'] == 1
second_mask = element['pclass'] == 2

element[first_mask | second_mask]
```

```python
#easier way 
element['pclass'].isin([1,2])
```
